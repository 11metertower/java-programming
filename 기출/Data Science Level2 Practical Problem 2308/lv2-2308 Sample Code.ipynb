```python
from sklearn.preprocessing import StandardScaler
scl00 = StandardScaler()
scl00.fit(df210.loc[:, 'AF3':'AF4'])
df_tmp = pd.DataFrame(scl00.transform(df210.loc[:,'AF3':'AF4']), columns=df210.loc[:, 'AF3':'AF4'].columns)
df211 = pd.concat([df210['PTC_ID'], df_tmp, df210['EYE_CLOSED']], axis=1)
```

```python
df211.describe()
```

1-2

```python
list_cols = list(df210.loc[:, 'AF3':'AF4'].columns)
df_tmp00 = df211[df211['EYE_CLOSED']==0].copy()
df_tmp01 = df211[df211['EYE_CLOSED']==1].copy()
dic_tmp00 = {}
for col in list_cols:
    dic_tmp00[col] = np.abs(df_tmp00[col].mean() - df_tmp01[col].mean())

srs_tmp00 = pd.Series(dic_tmp00).sort_values(ascending=False)
list_03 = list(srs_tmp00.iloc[:3].index)
list_11 = list(srs_tmp00[3:].index)
```

```python
srs_tmp00
```

1-3

```python
df212 = df211[df211['EYE_CLOSED']==1].copy()
df212.shape
```

1-4

```python
list_11
```

```python
list_col00 = []
list_col01 = []
list_mean = []
for col00 in list_03:
    for col01 in list_11:
        arr_tmp00 = np.zeros(19)
        arr_tmp01 = np.zeros(19)
        srs_tmp00 = df212[col00].copy()
        srs_tmp01 = df212[col01].copy()
        for i, num_q in enumerate(np.arange(0.05, 1, 0.05)):
            arr_tmp00[i] = srs_tmp00.quantile(num_q)
            arr_tmp01[i] = srs_tmp01.quantile(num_q)
            print(num_q)
        num_mean00 = np.mean(np.abs(arr_tmp00 - arr_tmp01))
        list_col00.append(col00)
        list_col01.append(col01)
        list_mean.append(num_mean00)
df_tmp00 = pd.DataFrame({'var03':list_col00, 'var11':list_col01, 'mean':list_mean})
```

1-5

```python
str_col = df_tmp00.sort_values('mean').iloc[0, 1]
res21 =  df_raw01[str_col].mean()
print('res21 : ', res21)
ans21 = round(res21, 2)
print('ans21 : ', round(res21, 2))
```

```python
df220 = df_raw01.copy()
```

2-1

```python
dic_tmp = {}
list_cols = list(df220.loc[:, 'AF3':'AF4'].columns)
for col in list_cols:
    num_std = df220[col].std()
    num_mean = df220[col].mean()
    num_cv = num_std/num_mean
    dic_tmp[col] = [num_cv]
srs_tmp = pd.DataFrame(dic_tmp).T

list_05 = list(srs_tmp.sort_values(0, ascending=False)[:5].index)
df221 = df220.loc[:, list_05].copy()
```

```python
df222 = df221.copy()
for col in df221.columns:
    num_q1 = df221[col].quantile(0.25)
    num_q3 = df221[col].quantile(0.75)
    num_iqx = num_q3 - num_q1
    num_q2 = df221[col].quantile(0.5)
    df222 = df222.loc[(df222[col]<=num_q3 + 1.5*num_iqx) & \
                      (df222[col]>=num_q1 - 1.5*num_iqx), :].copy()
    print(col," : ",df222.shape[0])
```

2-3

```python
for col in df222.columns:
    df222[col] = (df222[col] - df222[col].min())/(df222[col].max() - df222[col].min())
df222.describe()
```

2-4

```python
from time import time
```

```python
srs_y = df220.loc[df222.index, 'EYE_CLOSED'].copy()
dic_tmp = {}

time00 = time()

from sklearn.cluster import KMeans
km00 = KMeans(random_state=1234, n_clusters=2, n_init=150)  # n_init=100 추가
for col in df222.columns:
    km00.fit(df222.drop([col], axis=1))
    srs_cst = pd.Series(km00.labels_)
    srs_cst.index = df222.index
    num_mean00 = srs_y[srs_cst==0].mean()
    num_freq00 = (srs_cst==0).sum()
    dic_tmp[col+'00'] = [num_mean00,num_freq00]
    num_mean01 = srs_y[srs_cst==1].mean()
    num_freq01 = (srs_cst==1).sum()
    dic_tmp[col+'01'] = [num_mean01,num_freq01]

print("소요시간 : ", time() - time00) 
    
df_tmp00 = pd.DataFrame(dic_tmp).T
df_tmp00
```

```python
res22 = df_tmp00.loc[df_tmp00[0]==df_tmp00[0].max(), 1].values[0]/df222.shape[0]
ans22 = round(res22,2)
print('res22 : ',res22)
print('ans22 : ',ans22)
```

```python
df230 = df_raw01.copy()
```

3-1

```python
train230 = df230[df230['PTC_ID'] % 5 !=0].copy()
test230 = df230[df230['PTC_ID'] % 5 ==0].copy()
print(train230.shape, test230.shape)
```

3-2

```python
list_14 = list(train230.loc[:, 'AF3':'AF4'].columns)
```

```python
from sklearn.linear_model import LinearRegression
list_rsq = []
for col in list_14:
    LR00 = LinearRegression()
    list_tmp = list_14.copy()
    list_tmp.remove(col)
    LR00.fit(train230.loc[:, list_tmp], train230[col])
    arr_pred = LR00.predict(train230.loc[:, list_tmp])
    arr_act = np.array(train230[col])
    arr_mean = np.mean(arr_act)
    num_rsq = 1 - np.sum((arr_pred - arr_act)**2)/np.sum((arr_pred - arr_mean)**2)
    list_rsq.append(num_rsq)
```

```python
df_tmp = pd.DataFrame({'col':list_14, 'rsq':list_rsq})
list_06 = list(df_tmp.loc[df_tmp['rsq']<=0.95, 'col'])
```

```python
list_06
```

```python
from sklearn.linear_model import LogisticRegression
LR00 = LogisticRegression(solver = 'saga',max_iter=1000 , C = 100000, random_state = 1234)
LR00.fit(train230.loc[:, list_06], train230['EYE_CLOSED'])
arr_pred = LR00.predict(test230.loc[:, list_06])
arr_act = np.array(test230['EYE_CLOSED'])
```
