```python
import pandas as pd
import numpy as np
```

```python
raw00 = pd.read_csv('lv2-2406.csv')
raw00.info()
```

# 1

```python
df210 = raw00.copy()
```

1-1

```python
from math import log2

list_ent = []
for col in ['PART02','PART03','PART04','PART05','PART06']:
    e01 = 0
    for i in ['A','B']:
        e00 = 0
        for j in ['G','B']:
            p00 = np.sum((df210[col]==i) & (df210['CLASS']==j))/np.sum(df210[col]==i)
            e00 = e00 - p00*log2(p00)
        e01 = e01 + (np.sum(df210[col]==i)/df210.shape[0])*e00    
    list_ent.append(e01)
list_ent
```

1-2

```python
res21 = np.min(list_ent) + np.max(list_ent)
ans21 = round(res21, 2)
print(res21)
print('ans1 : ', ans21)
```

# 2

```python
df220 = raw00.copy()
```

2-1

```python
list_cols = ['PART00','PART01','PART07','PART08','PART09']
df_dum = pd.get_dummies(df220.loc[:, list_cols], drop_first=True)
df221 = pd.concat([df220['CLASS'], df_dum], axis=1)
list_dum = list(df_dum.columns)
df221
```

```python
len(list_dum)
```

2-2

```python
list_inf = []
for col in list_dum:
    pa0 = np.sum((df221[col]==0) & (df221['CLASS']=='G'))/np.sum(df221[col]==0)
    pa1 = np.sum((df221[col]==1) & (df221['CLASS']=='G'))/np.sum(df221[col]==1)
    val00 = np.max([pa0/(pa1+0.01), pa1/(pa0+0.01)])
    list_inf.append(val00)
list_inf
```

```python
res22 = np.mean(np.sort(list_inf)[-10:])
ans22 = round(res22, 3)
print(res22)
print('ans2 : ', ans22) 
```

# 3

```python
df230 = raw00.copy()
```

3-1

```python
list_cols = ['PART10','PART11','PART12','PART13']
df_dum = pd.get_dummies(df230.loc[:, list_cols], drop_first=True)
df231 = pd.concat([df230.loc[:, ['PRD_ID','CLASS']], df_dum], axis=1)
list_dum = list(df_dum.columns)
df231
```

```python
len(list_dum)
```

3-2

```python
train230 = df231[df231['PRD_ID']%5!=0].copy()
test230 = df231[df231['PRD_ID']%5==0].copy()
print(train230.shape, test230.shape)
```

3-3

```python
from sklearn.tree import DecisionTreeClassifier
DT00 = DecisionTreeClassifier(random_state=1234, max_depth=9)
DT00.fit(train230.loc[:, list_dum], train230['CLASS'])
```

3-4

```python
srs_col = pd.Series(list_dum)
srs_imp = pd.Series(DT00.feature_importances_)
df_imp = pd.concat([srs_col, srs_imp], axis=1)
df_imp = df_imp.sort_values(1, ascending=False)
df_imp.columns = ['feat', 'imp']
df_imp.index = range(df_imp.shape[0])
df_imp
```

3-5

```python
list_acc = []
for i in [5, 10, 14]:
    DT01 = DecisionTreeClassifier(random_state=1234, max_depth=9)
    list_vars = list(df_imp.iloc[:i, 0])
    DT01.fit(train230.loc[:, list_vars], train230['CLASS'])
    arr_pred = DT01.predict(test230.loc[:, list_vars])
    arr_act = np.array(test230['CLASS'])
    acc00 = np.sum(arr_pred==arr_act)/arr_act.shape[0]
    list_acc.append(acc00)
res23 = np.mean(list_acc)
ans23 = round(res23, 3)
print(res23)
print('ans3 : ', ans23)
```
